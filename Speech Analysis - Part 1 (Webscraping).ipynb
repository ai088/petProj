```python
# import libraries
import pandas as pd
import requests
import bs4
import re
```

```python
url = 'https://www.rev.com/blog/transcripts/david-attenborough-cop26-climate-summit-glasgow-speech-transcript'
```

```python
print(url)
```

```python
# make test resquest
test_res = requests.get(url)
test_res
```

```python
soup = bs4.BeautifulSoup(test_res.text,'html')
soup
```

Script body is found in '<'div class="fl-callout-text"'>'

```python
script = soup.find('div', {"class":"fl-callout-text"})
script
```

```python
script.find_all('p')
```

```python
Lst = []
for p in script.find_all('p'):
    script_dict = {}
    script_dict['text'] = p.text
    Lst.append(script_dict)
df = pd.DataFrame(Lst)
df
```

```python
# review
df['text'][0]
```

```python
# trim and get only the speakers transcript we need
sub_df = df.drop(index=[3,4,5,8,9,10])
```

```python
# reset index
sub_df.reset_index(drop=True, inplace=True)
```

```python
# what we need is actually the latter part of text separated by newline character
sub_df['text'][0].split('\n')[1]
```

```python
sub_df['cleaned_text'] = sub_df['text'].apply(lambda x: x.split('\n')[1])
```

```python
# review the data
sub_df
```

```python
# Replace ?/! with .
specialchars = '[?!]'
# create an interim column of speech after replacement
sub_df['cleaned_text'] = sub_df['cleaned_text'].apply(lambda x: re.sub(specialchars,'.',x))
```

```python
full_text = ' '.join(sub_df['cleaned_text'])
```

```python
# Write to text file
with open("COP26speech.txt", mode = "w") as f:
    f.write(full_text)
```
